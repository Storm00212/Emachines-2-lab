```python
data=pd.read_csv("/content/table.csv")
data.head()
```

```python
#tranformaing value of current and voltage into rms
data['current']=data['current']/np.sqrt(2)
```

```python
data.head()
```

```python
plt.title("Supply voltage vs current at no load")
plt.plot(data.current,data.voltage,marker='o',label="No load")
plt.legend()
plt.xlabel("Current")
plt.ylabel("Supply Voltage")
plt.grid(True)
```

```python
apparent_power=data.current*data.voltage
data['PowerFactor']=data.power/apparent_power
data.PowerFactor
```
